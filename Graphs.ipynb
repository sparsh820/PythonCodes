```python
import queue
class Graph:
    def __init__(self,nVertices):
        self.nVertices=nVertices
        self.adjMatrix=[[0 for j in range(nVertices)] for i in range(nVertices)]
    def addEdge(self,v1,v2):
        self.adjMatrix[v1][v2]=1
        self.adjMatrix[v2][v1]=1
        
    def removeEdge(self,v1,v2):
        if self.containsEdge(v1,v2) is False:
            return
        self.adjMatrix[v1][v2]=0
        self.adjMatrix[v2][v1]=0
        
    def containsEdge(self,v1,v2):
        return True if self.adjMatrix[v1][v2]>0 else False
    
    def __dfsHelper(self,sv,visited):
        print(sv)
        visited[sv]=True
        for i in range(self.nVertices):
            if self.adjMatrix[sv][i]>0 and visited[i] is False:
                self.__dfsHelper(i,visited)
                
    def dfs(self):
        visited=[False for i in range(self.nVertices)]
        for i in range(self.nVertices):
            if visited[i]==False:
                self.__dfsHelper(i,visited)
        
    def __bfsHelper(self,sv,visited):
        q=queue.Queue()
        q.put(sv)
        visited[sv]=True
        while q.empty() is False:
            u=q.get()
            print(u,end=' ')
            for i in range(self.nVertices):
                if self.adjMatrix[u][i]>0 and visited[i] is False:
                    q.put(i)
                    visited[i]=True
                    
        
    def bfs(self):
        visited=[False for i in range(self.nVertices)]
        for i in range(self.nVertices):
            if visited[i]==False:
                self.__bfsHelper(i,visited)
    
    def __HasPath(self,v1,v2,visited):
        if self.containsEdge(v1,v2)==True:
            return True
        visited[v1]=True
        for i in range(self.nVertices):
            if self.adjMatrix[v1][i]>0 and self.__HasPath(i,v2,visited):
                return True
        return False
    
    def HasPath(self,v1,v2):
        visited=[False for i in range(self.nVertices)]
        return self.__HasPath(v1,v2,visited)
    
    def __getPathDFS(self, sv, ev, visited) :
        if sv == ev :
            return [sv]
        visited[sv] = True
        for i in range(self.nVertices) :
            if self.adjMatrix[sv][i] == 1 and not visited[i] :
                li = self.__getPathDFS(i, ev, visited)
                if li != None :
                    li.append(sv)
                    return li
        return None
    
    def getPathDFS(self, sv, ev) :
        visited = [False for i in range(self.nVertices)]
        return self.__getPathDFS(sv, ev, visited)
    
    def __getPathBFS(self, sv, ev, visited) :
        mapp = {} #child-parent map
        q = queue.Queue()
        if self.adjMatrix[sv][ev] == 1 and sv == ev :
            ans = []
            ans.append(sv)
            return ans
        q.put(sv)
        visited[sv] = True
        while q.empty() is False :
            front = q.get()   
            for i in range(self.nVertices) :
                if self.adjMatrix[front][i] == 1 and visited[i] is False :
                    mapp[i] = front
                    q.put(i)
                    visited[i] = True
                    print(mapp)
                    if i == ev :
                        ans = []
                        ans.append(ev)
                        value = mapp[ev]
                        
                        while value != sv :
                            ans.append(value)
                            value = mapp[value]
                        
                        ans.append(value)
                        return ans
    
        return []
        
        
    def getPathBFS(self, sv, ev) :
        
        # Return empty list in case sv or ev is invalid
        if (sv > (self.nVertices - 1)) or (ev > (self.nVertices - 1)) :
            return list()
        visited = [False for i in range(self.nVertices)]
        return self.__getPathBFS(sv, ev, visited)
    
    def isConnectedHelper(self,sv,visited):
        visited[sv] = True
        for i in range(self.nVertices) :
            if self.adjMatrix[sv][i] == 1 and not visited[i] :
                self.isConnectedHelper(i,visited)   
                
    def isConnected(self) :
        visited = [False for i in range(self.nVertices)]
        self.isConnectedHelper(0,visited)
        for boolVal in visited :
            if not boolVal:
                return False
        return True
    def cchelper(self,sv,visited,smalloutput):
        q=queue.Queue()
        q.put(sv)
        visited[sv]=True
        while not q.empty():
            u=q.get()
            smalloutput.append(u)
            print(smalloutput)
            for i in range(self.nVertices):
                if self.adjMatrix[u][i]==1 and visited[i] is False:
                    q.put(i)
                    visited[i]=True
            
    
    
    def allconnectedComponents(self):
        visited=[False for i in range(self.nVertices)]
        output=[]
        for i in range(len(visited)):
            if not visited[i]:
                smalloutput=[]
                self.cchelper(i,visited,smalloutput)
                output.append(smalloutput)
        return output
    
    def __str__(self):
        return str(self.adjMatrix)
    
      
```

```python
def getParent(v,parent):
    if v == parent[v]:
        return v
    return getParent(parent[v],parent)

def kruskal(edges,V,E):
    edges = sorted(edges,key = lambda edge:edge.wt)
    output=[] #mst
    parent= [i for i in range(V)] #sets initially set to parent upon themselves
    count=0
    i=0
    print(parent)
    while count<V-1:
        currentEdge=edges[i]
        srcParent=getParent(currentEdge.src,parent)
        destParent=getParent(currentEdge.dst,parent)
        if srcParent != destParent:
            output.append(currentEdge)
            count+=1
            parent[srcParent] = destParent
        i+=1
    print(parent)
    return output
```

```python
class Solution:
    
    #Function to detect cycle in a directed graph.
    def isCyclic(self, V, adj):
        visited = [0]*V
        pathTraversed = [0]*V
        
        def dfs(node):
            visited[node] = 1
            pathTraversed[node] = 1
            for neighbour in adj[node]:
                
                if not visited[neighbour] and dfs(neighbour): return True
                if pathTraversed[neighbour]: return True
                
            pathTraversed[node] = 0
            return False
            
        for node in range(V):
            if not visited[node]:
                if dfs(node): return 1
                
        return 0
```

```python

class Solution:
    
    #Function to return a list containing the DFS traversal of the graph.
    def dfsOfGraph(self, V, adj):
        visited=[False]*V
        ans=[]
        def dfs(node):
            ans.append(node)
            visited[node]=True
            for nbr in adj[node]:
                if not visited[nbr]:
                    dfs(nbr)
        for node in range(V):
            if not visited[node]:
                dfs(node)
        return ans
```

```python
import queue
class Solution:
    
    #Function to return Breadth First Traversal of given graph.
    def bfsOfGraph(self, V, adj):
        visited=[False]*V
        ans=[]
        def bfs(node):
            q=queue.Queue()
            q.put(node)
            visited[node]=True
            while not q.empty():
                rn=q.get() #removed node
                ans.append(rn)
                for nbr in adj[rn]:
                    if not visited[nbr]:
                        q.put(nbr)
                        visited[nbr]=True
                    
        bfs(0)
        return ans
                    
```
